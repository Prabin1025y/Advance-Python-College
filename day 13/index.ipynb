### CHUNKING

```python
import random
import pandas as pd
import tqdm

# create a very big csv file
names=["ram", 'shyam', 'hari', 'krishna', 'karan', 'sita']
final_list = []
for i in tqdm.tqdm(range(1000000)):
    random_name = names[random.randint(0, 5)]
    final_list.append({
        "Emp_id": i + 1,
        "Name": random_name,
        "salary": 100000 * random.random()
    })
df = pd.DataFrame(final_list)
df.to_csv("BigData.csv", index=False)
    
```

```python
first_chunk = True
output_file = 'processed_data.csv'
for chunk in pd.read_csv("BigData.csv", chunksize=100000):
    chunk['Gender'] = chunk['Name'].apply(lambda x: 'Female' if x == 'sita' else 'Male')
    chunk.to_csv(
        output_file,
          mode = 'w' if first_chunk else 'a',
          header=first_chunk,
          index='False'
          )
    first_chunk = False
```
