```python
import requests

cities = ['Dharan', 'Damak', 'Butwal', 'Pokhara', 'Biratnagar']
api_key = "1e679f6aa64bfb3ba0eabd0dff937661"

new_df = pd.DataFrame(columns=[
    'name',
    'main.temp',
    'main.humidity',
    'wind.speed',
    'main.feels_like',
    'Description'
])

for city in cities:
    url=f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"

    response = requests.get(url)
    # print(response.status_code)
    if(response.status_code == 200):
        data = response.json()
        df = json_normalize(data)
        df_weather= df[[
            'name',
            'main.temp',
            'main.humidity',
            'wind.speed',
            'main.feels_like',
            'weather'
        ]].copy()
        df_weather['Description'] = df['weather'].apply(lambda x: x[0]['description'])
        df_weather = df_weather.drop(columns=['weather'])
        new_df = pd.concat([new_df, df_weather], ignore_index=True)

print(new_df)
```
