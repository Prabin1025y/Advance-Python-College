```python
import matplotlib.pyplot as plt
import numpy as np

X = np.arange(1, 10, 0.2)
y = np.square(X)

plt.plot(X, y, c = 'r', alpha=0.7)
plt.xlabel("X")
plt.ylabel("y = X^2")
plt.title("Sales over time")
plt.legend()
plt.show()
```

```python
categories = ['Apple', 'Bananas', 'Cherries', 'Dates']
values = [10, 25, 17, 12]
colors = ['red', 'gold', 'purple', 'brown']

# plt.bar(categories, values, width=0.8, edgecolor='black')
plt.barh(categories, values, color=colors)

plt.xlabel('Fruit')
plt.ylabel('Quantity')
plt.title('Fruit Inventory')

plt.show()
```

```python
import seaborn as sns

data = np.random.normal(loc = 50, scale = 10, size = 1000)

sns.histplot(data, bins=20, color='skyblue', edgecolor='black', kde=True)
plt.xlabel("Vlaue")
plt.ylabel("Frequiency")
plt.title("Histogram of Random Data")
plt.show()
```

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(
    n_features=2, n_samples=1000, random_state=42
)
print(X.shape)
print(y.shape)
```

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap='coolwarm', edgecolors='black')
plt.show()
```
