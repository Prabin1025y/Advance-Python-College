# Collections

```python
from collections import Counter

nums = ["abracadabra", 1, 3, 4, 4, 4, 5]
c = Counter(nums)

print(c)
print(type(c))
```

```python
from collections import defaultdict
aDict = defaultdict(int)
aDict["a"] = aDict["a"] + 1
print(aDict)
```

```python
from collections import deque
q = deque([1,2,3,4])
print(q)
q.appendleft(100)
print(q)
q.append(200)
print(q)
q.pop()
print(q)
q.popleft()
print(q)
q.clear()
print(q)
```

```python
from collections import namedtuple
Point = namedtuple("Point", 'x y')
p = Point(3,4)
print(p)
```
