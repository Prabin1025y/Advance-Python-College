```python
# wap to compute factorial using while loop
def fact(n):
    i = 1
    fact = 1
    while i <= n:
        fact *= i
        i += 1
    return fact
```

```python
print(fact(5))
```

```python
# *args
def average(*args):
    s = 0
    for i in args:
        s += i
    return s / len(args)

average(1,2,3)
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __str__(self):
        return f'({self.x}, {self.y})'
    
    def __add__(self, other):
        new_x = self.x +  other.x
        new_y = self.y + other.y
        return Point(new_x, new_y)
    
p1 = Point(2,3)
p2 = Point(3,2)
p3 = p1 + p2 
print(p3)
```

```python
#HomeWork
#WAP to create a class named Distance that has feet and inches as data members, Overload any 2 operator\

class Distance:
    def __init__(self, feet, inches):
        self.feet = feet
        self.inches = inches

        while self.inches >= 12:
            self.feet += 1
            self.inches -= 12
    
    def __str__(self):
        return f"{self.feet} feet {self.inches} inches."

    
    def __add__(self, other):
        new_feet = self.feet + other.feet
        new_inches = self.inches + other.inches
        while new_inches >= 12:
            new_feet += 1
            new_inches -= 12
        return Distance(new_feet, new_inches)
    
    def __mul__(self, scale):
        new_feet = self.feet * scale
        new_inches = self.inches * scale

        while new_inches >= 12:
            new_feet += 1
            new_inches -= 12
        
        return Distance(new_feet, new_inches)


d1 = Distance(5, 11)
d2 = Distance(6, 4)
d3 = d1 + d2
d4 = d1 * 2
print(d3)
print(d4)
```
