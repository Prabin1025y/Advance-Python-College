```python
aDict={}
with open(file='text.txt', mode='r') as f:
    content = f.read().lower()
    for word in content.split():
        if word not in aDict:
            aDict[word] = 1
        else:
            aDict[word] += 1

print(aDict)
    
    
```

```python
with open(file='text.txt', mode = 'r') as f:
    content = f.readlines()
    for line in content:
        print(line, end='')
```

```python
with open(file='new.txt',  mode='w') as f:
    while True:
        input_text = input("Enter some text: ")
        if input_text == 'quit':
            break
        f.write(input_text)
        f.write("\n")
```

```python
import csv

data = [
    ["Name", "Address", "age"],
    ["Prabin", "Gauradaha", "22"],
    ["Suman", "Pokhara", "20"],
    ["Karan", "CharKoseJhadi", "60"],
]
with open(file="exampleWrite.csv", mode="w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(data)
```

```python
import csv

data = [
    {"Name": "Prabin", "Address": "Gauradaha", "Age": "22"},
    {"Name": "Suman", "Address": "Pokhara", "Age": "20"},
    {"Name": "Karan", "Address": "CharKoseJhadi", "Age": "60"},
]
with open(file="exampleWriteDict.csv", mode="w", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=['Name','Address','Age'])
    writer.writeheader()
    writer.writerows(data)
```
