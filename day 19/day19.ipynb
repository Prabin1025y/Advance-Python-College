```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
ax1.plot(np.random.standard_normal(50).cumsum(), color='black', linestyle='dashed');
ax2.hist(np.random.standard_normal(100), bins=20, color='black', alpha=0.3);
ax3.scatter(np.arange(30), np.arange(30)+3*np.random.standard_normal(30));
ax4.bar(x=['A', 'B', 'C', 'D', 'E'], height=[25, 13, 21, 13, 2]);
```

```python
fig, axes = plt.subplots(2,2, sharex=True, sharey=True)

for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.standard_normal(500), bins = 50, color='black', alpha = 0.5)
fig.subplots_adjust(wspace=0, hspace=0)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(np.random.standard_normal(1000).cumsum())
ticks=ax.set_xticks([0, 250, 500, 750, 1000])
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=30, fontsize=10)
ax.set_title("My Plot")
ax.text(x= 10, y=10, s="Hello World", family='monospace', fontsize=10);
```

```python
x = np.arange(1,6)
y = x * 2
data = np.random.randn(100)

fig, axs = plt.subplots(2,2, figsize=(6,4))

axs[0,0].plot(x,y)
axs[0,0].set_title("Line Plot")

axs[0,1].bar(x,y)
axs[0,1].set_title("Bar Plot")


axs[1,0].hist(data)
axs[1,0].set_title("Histogram")

axs[1,1].scatter(x,y)
axs[1,1].set_title("Scatter Plot")

plt.tight_layout()

```
