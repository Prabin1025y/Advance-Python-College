### Question 1: Write a Python program to check if a given number is prime

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n / 2) + 1):
        if n % i == 0:
            return False
    return True


num = int(input("Enter a number: "))
if is_prime(num):
    print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")
```

### Question 2: Create a function that converts a list of temperatures from Celsius to Fahrenheit

```python
def celsius_to_fahrenheit(celsius_list):
    return [(c * 9 / 5) + 32 for c in celsius_list]


temps_celsius = [0, 20, 37, 100]
temps_fahrenheit = celsius_to_fahrenheit(temps_celsius)
print(temps_fahrenheit)
```

### Question 3: Write a program to count the number of vowels in a given string

```python
def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0

    for char in text:
        if char in vowels:
            count += 1

    return count


string = input("Enter a string: ")
print("Number of vowels:", count_vowels(string))
```

### Question 4: Implement a function to find the factorial of a number using recursion

```python
def factorial(n):
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers")
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)


num = int(input("Enter a number: "))
print("Factorial:", factorial(num))
```

### Write a Python script to merge two dictionaries into one

```python
def merge_dictionaries(dict1, dict2):
    merged = dict1.copy()
    merged.update(dict2)
    return merged


dict_a = {"a": 1, "b": 2}
dict_b = {"c": 3, "d": 4}

result = merge_dictionaries(dict_a, dict_b)
print(result)
```

### Question 6: Create a function that takes a list of numbers and returns only the even numbers

```python
def get_even_numbers(numbers):
    return [num for num in numbers if num % 2 == 0]


nums = [1, 2, 3, 4, 5, 6, 7, 8]
print(get_even_numbers(nums))
```

### Question 7: Write a program to check if a string is a palindrome

```python
def is_palindrome(text):
    cleaned_text = text.lower().replace(" ", "")
    return cleaned_text == cleaned_text[::-1]


string = input("Enter a string: ")
if is_palindrome(string):
    print("The string is a palindrome.")
else:
    print("The string is not a palindrome.")
```

### 8. Implement a class Circle with a method to calculate the area and circumference.

```python
import math

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

    def circumference(self):
        return 2 * math.pi * self.radius
```

### 9. Write a Python function that reads a text file and counts the number of lines

```python
def count_lines(filename):
    with open(filename, "r") as file:
        return sum(1 for _ in file)

print(count_lines("file1.txt"))
```

### 10.Create a program to sort a list of tuples based on the second element.

```python
def sort_by_second(tuples_list):
    return sorted(tuples_list, key=lambda x: x[1])

aTupleList = [(50, 10, 15), (100, 2, 3), (0, 1, 5)]
sortedTuple = sort_by_second(aTupleList)
print(sortedTuple)
```

### 11.Implement a function to remove duplicates from a list without using set().


```python
def remove_duplicates(lst):
    result = []
    for item in lst:
        if item not in result:
            result.append(item)
    return result

aList= [1,2,3,3,4,4,4,5]
l2 = remove_duplicates(aList)
print(l2)
```

### 13.Create a class Rectangle with methods to calculate area and perimeter.


```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)
```

### 14.Write a function to reverse a list without using the built-in reverse() method.


```python
def reverse_list(lst):
    reversed_list = []
    for item in lst:
        reversed_list.insert(0, item)
    return reversed_list

aList= [1,2,3,3,4,4,4,5]
reversedList = reverse_list(aList)
print(reversedList)
```

### 15.Implement a Python function that generates the Fibonacci sequence up to n terms


```python
def fibonacci(n):
    sequence = []
    a, b = 0, 1

    for _ in range(n):
        sequence.append(a)
        a, b = b, a + b

    return sequence

print(fibonacci(10))
```

### 16.Write a program to count the frequency of each word in a given string.


```python
def word_frequency(text):
    freq = {}
    for word in text.split():
        freq[word] = freq.get(word, 0) + 1
    return freq

print(word_frequency("Hello world. Hello boy, how are you"))
```

### 19.Write a Python program to check if two strings are anagrams of each other.
Anagram means, if two word contains same letters in same quantity but they can be in different orders.


```python
def are_anagrams(str1, str2):
    return sorted(str1.lower()) == sorted(str2.lower())

print(are_anagrams("study", "dusty"))
```

### 20.Create a function that calculates the sum of digits of a given number.

```python
def sum_of_digits(number):
    return sum(int(digit) for digit in str(abs(number)))

print(sum_of_digits(343434))
```
