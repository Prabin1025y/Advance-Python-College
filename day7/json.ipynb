# Reading a file

```python
import json

from pprint import pprint
with open(file="data.json", mode='r') as f:
    content = json.load(f)

pprint(content)
```

# Writing into json file

```python
import json

aDict={"name": "Prabin Acharya", "Address":"Gauradaha" }

with open(file='write.json', mode='w') as f:
    json.dump(aDict, f, indent=4)
```
