## Covariance using numpy and pandas

```python
import numpy as np
import pandas as pd

x = np.array([10, 20, 30, 40, 50])
y = np.array([10, 9, 8, 7, 1])

np.cov(x,y)
```

```python
df = pd.DataFrame({"X": x, "Y": y})
df.cov()
```

```python
np.corrcoef(x, y)
```

```python
df.corr()
```
