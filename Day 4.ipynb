# Iterators

```python
aList = [1,2,3,4]
it = iter(aList)
print(it)
print(type(it))
```

```python
print(next(it))
print(next(it))
print(next(it))
print(next(it))
print(next(it))
```

```python
for num in aList:
    print(num)
```

```python
aList = [1,2,3,4]
it = iter(aList)
while True:
    try:
        num = next(it)
        print(num)
    except StopIteration:
        break
```

# Generators

```python
def squares(x):
    for i in range(x):
        yield i*i


print(squares(5))
print(type(squares(5)))

gen = squares(100)

for i in range(10):
    square = iter(gen)
    print(next(square))
```

# Decorator

```python
def myDecorator(func):
    def wrapper():
        print("Before")
        func()
        print("After")
    return wrapper
```

```python

def hello():
    print("Hello!")

sayHello = myDecorator(hello)
sayHello()
```

```python
hello()
```

```python
import time

def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f"Elapsed Time is {end-start} seconds")
        return result
    return wrapper
```

```python
@timer
def fact(n):
    f = 1
    for i in range(1, n+1):
        f *= i
    print(f"Factorial of {n} is {f}")

fact(100)
```
